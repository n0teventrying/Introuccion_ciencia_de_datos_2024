```python
#Antes de comenzar importamos las librerias necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Lee aquí tu dataset
df = pd.read_csv('wine_quality_red-clean.csv',index_col=0)
df.head()
# Es recomendable realizar una pequeña exploración antes de pasar a la siguiente sección
```

```python
#Bueno, no hay datos categoricos por loq que no ocupamos droppear a nadie
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, linewidths=.5);
```
